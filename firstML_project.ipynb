# **Load Data**

````python
import pandas as pd
df = pd.read_csv ('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
````

# **Data preparation**

## Data seperation as X and y

````python
y = df ['logS']
x = df.drop ('logS', axis=1)
````

## Data splitting

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
````

# **Model Building**

## **Linear regression**

**Training the model**

````python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
````

**Applying the model to make a prediction**

````python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
````

````python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
plt.scatter(x=y_lr_train_pred, y=y_train, c="#7CAE00", alpha=0.3)

# Add a line of best fit
z = np.polyfit(y_lr_train_pred, y_train, 1)
p = np.poly1d(z)
plt.plot(y_lr_train_pred, p(y_lr_train_pred), "#F8766D")

plt.ylabel('Experimental LogS')
plt.xlabel('Predicted LogS')
plt.show()
````

### Subtask:
Calculate and display evaluation metrics (e.g., R-squared, Mean Squared Error) for both the training and testing sets.


**Reasoning**:
Calculate and print the evaluation metrics for both training and testing sets.



````python
from sklearn.metrics import r2_score, mean_squared_error

r2_train = r2_score(y_train, y_lr_train_pred)
mse_train = mean_squared_error(y_train, y_lr_train_pred)
r2_test = r2_score(y_test, y_lr_test_pred)
mse_test = mean_squared_error(y_test, y_lr_test_pred)

print(f"Training set:")
print(f"R-squared: {r2_train:.2f}")
print(f"Mean Squared Error: {mse_train:.2f}")
print("\nTesting set:")
print(f"R-squared: {r2_test:.2f}")
print(f"Mean Squared Error: {mse_test:.2f}")
````
